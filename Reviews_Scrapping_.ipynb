```python
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('-headless')
options.add_argument('-no-sandbox')
options.add_argument('-disable-dev-shm-usage')
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import requests
import time
```

```python
def fetch_amazon_reviews_data(driver, url,value):
  comments = []
  ratings = []
  for i in range(51,100):
    driver.get(url + str (i))
    time.sleep(1)
    page = BeautifulSoup(driver.page_source,'lxml')
    all_divs = page.findAll('div')
    for ele in all_divs:
      if(ele.get('id')):
        if('-review-card' in ele.get('id')):
          get_span_data = ele.find('span',{'data-hook':'review-body'}).text
          if(ele.find('i',{'data-hook':'review-star-rating'})!= None):
            ratings.append(ele.find('i',{'data-hook':'review-star-rating'}).text.split(" ")[0])
            comments.append(get_span_data.replace('\n',''))
      next_button = driver.find_element_by_class_name("a-last")
  input_data = pd.DataFrame({'Name':value,'Reviews': comments,'Ratings':ratings})
  return input_data
```

```python
driver = webdriver.Chrome('chromedriver',options=options)
url = "https://www.amazon.com/Xiaomi-Band-Wristband-Magnetic-Bluetooth/product-reviews/B089NS9JW2/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber="
Xiaomi_Band_WristbandDF = fetch_amazon_reviews_data(driver,url,'Xiaomi-Band-Wristband-Magnetic-Bluetooth')
Xiaomi_Band_WristbandDF
```
